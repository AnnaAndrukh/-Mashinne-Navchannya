```python
#Завдання 1
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report
from keras.callbacks import EarlyStopping, ReduceLROnPlateau
```

```python
# 1. Завантажити датасет
fashion_mnist = keras.datasets.fashion_mnist
(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()
```

```python
# 2. Роздрукувати 10 зображень
plt.figure(figsize=(10,10))
for i in range(10):
    plt.subplot(1,10,i+1)
    plt.imshow(x_train[i], cmap='gray')
    plt.axis('off')
plt.show()
```

```python
# 3. Підготовка даних
x_train = x_train / 255.0
x_test = x_test / 255.0
y_train = keras.utils.to_categorical(y_train, 10)
y_test = keras.utils.to_categorical(y_test, 10)
```

```python
# 4. Побудова архітектури нейронної мережі
model = keras.Sequential([
    keras.layers.Input(shape=(28, 28)),
    keras.layers.Flatten(),
    keras.layers.Dense(128, activation='relu'),
    keras.layers.Dropout(0.2),
    keras.layers.Dense(10, activation='softmax')
])
```

```python
# 5. Використання ранньої зупинки і зменшення швидкості
early_stop = EarlyStopping(monitor='val_loss', patience=3)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=2)
```

```python
# 6. Компіляція моделі
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
# 7. Навчання моделі
history = model.fit(x_train, y_train, validation_data=(x_test, y_test), epochs=20, callbacks=[early_stop, reduce_lr])
```

```python
# 8. Виведення архітектури моделі
model.summary()
```

```python
# 10. Побудова графіків функції втрат і точності
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Test Loss')
plt.title('Функція втрат')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Test Accuracy')
plt.title('Точність')
plt.legend()
plt.show()
```

```python
# 11. Виведення класифікаційного звіту
y_pred = np.argmax(model.predict(x_test), axis=1)
y_true = np.argmax(y_test, axis=1)
print(classification_report(y_true, y_pred))
```

```python
# 12. Виведення прогнозованих і справжніх класів
for i in range(10):
    plt.imshow(x_test[i], cmap='gray')
    plt.title(f"Прогноз: {y_pred[i]}, Справжній: {y_true[i]}")
    plt.axis('off')
    plt.show()
```

```python
import keras_tuner as kt
```

```python
# 1. Функція для побудови моделі з можливістю тюнінгу гіперпараметрів
def build_model(hp):
    model = keras.Sequential()
    model.add(keras.layers.Flatten(input_shape=(28, 28)))
    hp_units = hp.Int('units', min_value=32, max_value=512, step=32)
    model.add(keras.layers.Dense(units=hp_units, activation='relu'))
    model.add(keras.layers.Dropout(0.2))
    model.add(keras.layers.Dense(10, activation='softmax'))

    hp_learning_rate = hp.Choice('learning_rate', values=[1e-2, 1e-3, 1e-4])
    model.compile(optimizer=keras.optimizers.Adam(learning_rate=hp_learning_rate),
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])
    return model
```

```python
# 2. Використання Hyperband для пошуку найкращих гіперпараметрів
tuner = kt.Hyperband(build_model,
                     objective='val_accuracy',
                     max_epochs=10,
                     factor=3)

tuner.search(x_train, y_train, epochs=10, validation_data=(x_test, y_test))
```

```python
# 3. Отримання найкращої моделі
best_model = tuner.get_best_models(num_models=1)[0]
best_model.summary()
```

```python
# 4. Порівняння результатів з завданням 1
best_model.evaluate(x_test, y_test)
```

```python
# 5. Виведення класифікаційного звіту
y_pred_best = np.argmax(best_model.predict(x_test), axis=1)
print(classification_report(y_true, y_pred_best))
```
